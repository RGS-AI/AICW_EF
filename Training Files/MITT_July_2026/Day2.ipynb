## OOP Concepts

```python
class Person:
    def __init__(self, name, gender, profession):
        self.name = name
        self.gender = gender
        self.profession = profession

    def show(self):
        print(f'Name: {self.name}, Gender:{self.gender} and Profession:{self.profession}')

    def work(self):
        print('Profession', self.profession)
```

```python
person = Person('RMS', 'M', 'Trainer')
```

```python
person.show()
```

```python
person.work()
```

```python
class person:
    def __init__(self, name, age, school):
        self.name = name
        self.age = age
        self.school = school
    def myfun(self):
        print("Hello, my name is ", self.name)
        print("My age is ", self.age)
        print("My school name is ", self.school)
```

```python
p1 = person('Sonu', 20, 'MIT')
p2 = person('Rakshitha', 21, 'MIT')
p3 = person('Sonali', 21, 'MIT')
```

### End-to-End Story Telling project on real-world dataset

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load the dataset
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# stats
df.describe()
```

```python
# no of rows and cols
df.shape
```

# Step 2: Data Cleaning

```python
# check for missing values
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.info()
```

# Step 3: Data Transformation

```python
# encode the variable - cat
df['gender'] = df['gender'].map({'male':0, 'female':1})
df['gender']
```

# Step 4: Data Visualization

```python
plt.subplot(221)

df['gender'].value_counts().plot(kind='bar', title='Gender of students', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['race/ethnicity'].value_counts().plot(kind='bar', title='Race/ethnicity of students')

plt.xticks(rotation=0)

plt.subplot(223)

df['lunch'].value_counts().plot(kind='bar', title='Lunch status of students')

plt.xticks(rotation=0)

plt.subplot(224)

df['test preparation course'].value_counts().plot(kind='bar', title='Test preparation course')

plt.xticks(rotation=0)

plt.show()
```

```python
df.boxplot()
```

```python
# Distribution of math scores
plt.figure(figsize=(10, 6))
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of math scores by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()
```

```python
# Scatter plot of reading score vs writing score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='reading score', y='writing score', hue='gender', data=df)
plt.title('Reading Score vs Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
# Heatmap of correlations
num_values = df.select_dtypes(include='number')
cm = num_values.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(cm, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
