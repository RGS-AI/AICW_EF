The price of a notebook is Rs. 45. A student buys 8 notebooks and gives a Rs. 500 note to the cashier. How much change will the student get back?

```python
price = 45
num_note_books = 8
cash = 500
# change - price * num = total price 
# cash - total price
total_price = num_note_books * price
change = cash - total_price
print(f'the total change for the transaction is {change}')
```
